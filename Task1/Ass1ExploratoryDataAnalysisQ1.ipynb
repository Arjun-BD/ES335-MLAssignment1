```python
import matplotlib.pyplot as plt
import numpy
import pandas as pd

fig, axs = plt.subplots(1, 6, figsize=(20,5), layout='constrained')

fig.suptitle("Activities Comparision")

def plot(fig, axs):
    activity = ["SITTING", "STANDING", "LAYING", "WALKING", "WALKING_DOWNSTAIRS", "WALKING_UPSTAIRS"]
    for i in range(len(axs)):
        sub1 = pd.read_csv(rf"../Combined/Train/{activity[i]}/Subject_3.csv")
        axs[i].plot(sub1)
        axs[i].legend(sub1.columns)
        axs[i].set_title(activity[i])
        
plot(fig, axs)
```
